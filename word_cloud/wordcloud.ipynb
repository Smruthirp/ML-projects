```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
text=["Smruthi","Ranjitha","Pavitra","Shreyas","Rakshitha","Cool","bffs","Love","together","Us",
      "Bitches","forever","bf_goals","Happy","Pretty","Insane","Crime_partners","my_people","Awesome",
      "favourites","besties","Always","Crazy","Fun"]
```

```python
word_cloud=WordCloud(width=600,height=500,background_color="white",max_words=100).generate(str(text))

fig=plt.figure(figsize=(10,10),facecolor=None)
plt.imshow(word_cloud)
plt.axis("off")
plt.show()
```
